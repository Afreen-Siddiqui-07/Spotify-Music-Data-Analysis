```python

df['duration_min'] = df['duration_ms'] / 60000
df['duration_min'] = df['duration_min'].round(2)
df.drop("duration_ms",inplace = True,axis=1)
```

```python
df.head()
```

```python
filtered_data = df.drop(columns=['key', 'mode'])
numeric_col = filtered_data.select_dtypes(include=['float64', 'int64']).columns

correlation_matrix = filtered_data[numeric_col].corr(method='pearson')
plt.figure(figsize=(16, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar=True)
plt.title("Correlation Heatmap Matrix ", fontsize=16)
plt.show()
```

```python

popularity_trend = df.groupby('year')['popularity'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=popularity_trend, x='year', y='popularity', marker='o', color="orange")
plt.title("Average Track Popularity Over the Years", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("popularity", fontsize=12)
plt.show()
```

```python
sns.set_palette("bright")
sns.set_style("whitegrid")


plt.figure(figsize=(10, 6))
sns.histplot(df['popularity'], bins=30, kde=True, color="teal")
plt.title("Distribution of Track Popularity", fontsize=16)
plt.xlabel("Popularity", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df, x='energy', y='danceability', 
    alpha=0.6, color="purple"
)
plt.title("Energy vs. Danceability", fontsize=16)
plt.xlabel("Energy", fontsize=12)
plt.ylabel("Danceability", fontsize=12)
plt.show()
```

```python

track_per_year = df['year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x=track_per_year.index, y=track_per_year.values, marker='o', color='green')
plt.title("No of Tracks Released Per Year", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("No. of Tracks", fontsize=12)
plt.show()
```
